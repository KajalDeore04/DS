# Feature engineering
- create new cols/ feat from existing data to extract insight

```python
# derived col

df = pd.DataFrame({
    'Length_cm': [100, 150, 200]
})
```

```python
# length_m col
df['Length_m'] = df['Length_cm'] /100
df
```

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 45, 65]
})
df
```

```python
# create age group

df['Age_Group'] = pd.cut(
                            df['Age'],
                             bins=[0,30,60,100],
                              labels=['Youth', 'Adult', 'Senior'] )
df
```

```python
# extract date parts

df = pd.DataFrame({
    'Date': pd.to_datetime(['2025-07-01', '2025-07-02'])
})

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df
```

## Mini Practice Project: Cleaning Titanic Dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

```python
df_clean = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df_filled = df.fillna('ffill')
df_filled.isnull().sum()
```

```python
df_filled['FamilySize'] = df_filled['SibSp'] + df_filled['Parch'] + 1
df_filled
```

```python
df_filled.dtypes
```

```python
df_filled['Age'].replace('ffill', 0, inplace=True)
df_filled['Age'].unique()
```

```python
df_filled['Age'].astype('int64')
df_filled.dtypes
```

```python
df_filled['IsMinor'] = df_filled['Age'].apply(lambda x: 1 if x < 18 else 0) 
```

```python
df_filled
```

```python
df_filled['Age'] = df_filled['Age'].astype('int64')
df_filled.dtypes
```

```python
df_filled['Title'] = df_filled['Name'].str.extract(',\s*([^\.]*)\s*\.', expand=False)
df_filled['Title'].value_counts()
```
