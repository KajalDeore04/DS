```python
import pandas as pd
import requests
from bs4 import BeautifulSoup # web scraping lib
```

```python
# requests.get('https://www.ambitionbox.com/list-of-companies?page=1').text

# ACCESS DENIED ERROR
# the website has a file where the pages are listed which should be prevented from the robots from accessing the data. if we hit a request like this the website considers us as a bot. 

# Instead of bot we have to disguise ourselves as a Browser request
# to do that we send a Header as well
 
```

```python
final = pd.DataFrame()

for i in range(1, 11):
    url = 'https://quotes.toscrape.com?page={}'.format(i)
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
    }
    webapage = requests.get(url, headers=headers).text

    soup = BeautifulSoup(webapage, 'lxml') # html parsing

    quote = []
    author = []
    tags = []

    for i in soup.find_all('div', class_='quote'):
        quote.append(i.find('span', class_='text').text)
        author.append(i.find('small', class_='author').text)
        
        tag_list = [tag.text for tag in i.find_all('a', class_='tag')]
        tags.append(', '.join(tag_list))  
    
    data = {
    'qoute': quote,
    'author': author,
    'tags' : pd.Series(tags)
    }

    df = pd.DataFrame(data)

    final = pd.concat([final, df], ignore_index=True)
```
