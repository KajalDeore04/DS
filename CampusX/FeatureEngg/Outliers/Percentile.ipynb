```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day44-outlier-detection-using-percentiles/weight-height.csv')
```

```python
sns.boxplot(x = df['Height'])
```

```python
# we decide an amount i.e. 99% or 1% and everything out of it is outlier

upper_limit = df['Height'].quantile(0.99)
lower_limit = df['Height'].quantile(0.01)
```

```python
# Capping --> Winsorization
new_df = df.copy()
new_df['Height'] = np.where(df['Height'] >= upper_limit,
        upper_limit,
        np.where(df['Height'] <= lower_limit,
        lower_limit,
        df['Height']))
```

```python
sns.boxplot(x = new_df['Height'])
```
