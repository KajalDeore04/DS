# Distribution Plots
- used to visualize distribution of quantitative data.(salary,age)
Types:
- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df
```

total_bill - quantitative

tip	- quantitative
	
size isnt quantitative coz has only 6 unique values out of 244 rows

```python
df['size'].unique()
```

# Distplot/histplot
- bins are created (range)
- datapoints are assigned if they lie in that particular bins
- kernel density tells us about skewness(kde= true) curve appears

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=20,kde=True )

plt.subplot(1,2,2)
sns.histplot(df['tip'], bins=20,kde=True )
```

# Joint Plot
- if u wanna compare both quantitative
- kind by default is scatter - dot appears at co-ords of data dot[18,19] and the line appearing on it is linear regression

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='reg')
```

# PairPlot
- plots all typesfor num data
- hue - helps differentiate between categorial data by colour

```python
sns.pairplot(df, hue='sex', palette='rainbow')
```

# RugPlot

```python
sns.rugplot(df)
```
