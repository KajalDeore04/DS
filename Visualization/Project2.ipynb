# Spotify Tracks Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:\Dataset-DS\spotify.csv")
df.head()
```

```python
df = df.dropna()
df.isna().sum()
```

```python
df.dtypes
```

## Cleaning and Setup

```python
df['duration_ms'] = df['duration_ms']/60000
df['duration_ms']
```

```python
df
```

1. What is the average duration, energy, and danceability of all songs?

```python
avg_duration = df['duration_ms'].mean()
print(avg_duration.round(2))

print(df['energy'].mean().round(2))

print(df['danceability'].mean().round(2))
```

2. Which artists have the most songs in the dataset?

```python
top_artist = df['artists'].value_counts().sort_values(ascending=False).head(1)
top_artist
```

3. Are there any duplicate tracks? Remove them.

```python
df['track_id'] = df['track_id'].drop_duplicates()
df
```

## Part 2 - visualization

### Track-Level Analysis

1. Top 10 artists with most tracks

```python
top_10_artists = df['artists'].value_counts().sort_values(ascending=False).head(10)

sns.barplot(x = top_10_artists.values, y=top_10_artists.index, palette='rainbow' )
```

2. Distribution of tempo, loudness, duration

```python
sns.histplot(data=df['tempo'], kde=True,  bins=20)
sns.histplot(data=df['loudness'], kde=True,  bins=20)
sns.histplot(data=df['duration_ms'], kde=True,  bins=20)
```

3. Compare energy across genres or years

```python
df['track_genre'].value_counts()
```

```python
top_10_genre = df['track_genre'].value_counts().sort_values(ascending=False).head(10).index

df_top_genre = df[df['track_genre'].isin(top_10_genre)]

# Plot boxplot of energy for each top genre
sns.boxplot(y='track_genre', x='energy', data=df_top_genre, palette='rainbow')
```

Energy vs Danceability (color = popularity)

```python
sns.scatterplot(x = df['energy'], y=df['danceability'], hue=df['popularity'])
```
