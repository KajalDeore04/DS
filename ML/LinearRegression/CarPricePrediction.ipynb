```python
# separate the input and output vars

X = df.drop(columns=['price'], axis=1)
y = df['price']
```

### Data cleaning and preprocessing

```python
# creates new columns for encoding
X_one_encoded = pd.get_dummies(X, columns=['model', 'transmission', 'fuelType'])
X_one_encoded = X_one_encoded.astype(int)
X_one_encoded
```

```python
# encodes inplace with 0,1,2,3,4 values

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

columns=['model', 'transmission', 'fuelType']

X_label_encoded = X

for i in columns:
    X_label_encoded[i] = encoder.fit_transform(X_label_encoded[i])


X_label_encoded
```

Normalization / Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
numerical_col = ['year','mileage','tax','mpg','engineSize']
```

```python
X_one_encoded[numerical_col] = scaler.fit_transform(X_one_encoded[numerical_col])
X_one_encoded
```

```python
# not everything is in 0-1 in label encoded data hence we need to scale all the columns

X_label_encoded[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(X_label_encoded[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
X_label_encoded
```

### Model Creation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

For One Hot Encoded

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_encoded, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

Evaluate

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - ((1 - r2)*(n - 1) / (n - p - 1))
adjusted_r2
```
