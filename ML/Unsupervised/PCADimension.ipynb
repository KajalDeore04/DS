```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
X, y = make_blobs(n_features=5, random_state=42, n_samples=500, centers=3, cluster_std=1.5)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)  # reduce no of feat to 2
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
df_pca = pd.DataFrame(X_pca, columns=['Feature1', 'Feature2'])
df_pca['label'] = y
```

```python
sns.scatterplot(data=df_pca, x='Feature1', y='Feature2',hue='label', palette='flare' )
```
