```python
import pandas as pd

df = pd.read_csv('C:/Dataset-DS/netflix_titles.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='%Y%m%d',errors='coerce')
```

```python
import matplotlib.pyplot as plt



plt.bar(df['type'].value_counts().index, df['type'].value_counts().values)
plt.show()
```

```python
plt.barh(df['country'].value_counts().head(5).index[::1],  df['country'].value_counts().head(5).values[::1] )
```

```python
plt.pie(df['rating'].value_counts().head(10).values,  labels=df['rating'].value_counts().head(10).index, autopct='%1.1f%%')

plt.show()
```
