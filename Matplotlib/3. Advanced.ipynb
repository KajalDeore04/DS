````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 7, 2, 5, 10]

plt.plot(x, y, marker='o')
plt.title("Annotated Line Plot")

# Add annotation at max value
plt.annotate(
    'Important Point',
    xy=(2, 7),
    xytext=(1.5, 8.5),
    arrowprops=dict(
        arrowstyle='fancy', 
        color='purple', 
        lw=2
    ),
    fontsize=12,
    color='purple',
    bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.3)
)


# Add text at a specific position
plt.text(1, 3, "Start", fontsize=10, color='blue')

plt.grid(True)
plt.show()
````

| Parameter    | Purpose                       |
| ------------ | ----------------------------- |
| `xy`         | Point to point at             |
| `xytext`     | Where text appears            |
| `arrowprops` | Dictionary to style the arrow |
| `fontsize`   | Size of text                  |
| `color`      | Text color                    |

## Twin Axes

````python
x = [1, 2, 3, 4, 5]
y1 = [3, 7, 2, 5, 10]
y2 = [4, 8, 5, 8, 15]


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()  # shares the same x-axis but different y-axis

ax1.plot(x, y1, color='blue')  # Plot 1 on left y-axis
ax2.plot(x, y2, color='green') # Plot 2 on right y-axis
````

````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
temp = [30, 32, 33, 35, 37]     # Temperature (C)
rain = [100, 80, 60, 90, 70]    # Rainfall (mm)

fig, ax1 = plt.subplots()

# Left Y-axis (temp)
ax1.plot(x, temp, color='red', marker='o')
ax1.set_xlabel('Day')
ax1.set_ylabel('Temperature (C)', color='red')

# Create second Y-axis sharing x-axis
ax2 = ax1.twinx()
ax2.plot(x, rain, color='blue', marker='s')
ax2.set_ylabel('Rainfall (mm)', color='blue')

plt.title("Twin Axes: Temperature vs Rainfall")
plt.grid(True)
plt.show()
````

# Logarithmic Scales
- if the data is too big on scale which doesnt fit the box then use logarithmic values as scale for labels
- You can set the x-axis, y-axis, or both to a log scale using:
    ```
    plt.xscale('log')
    plt.yscale('log')
    ```

    Or, if using Axes:
    ```
    ax.set_xscale('log')
    ax.set_yscale('log')
    ```


````python
import matplotlib.pyplot as plt
import numpy as np

x = np.logspace(0.1, 2, 100)
y = x**2

plt.plot(x, y)
plt.xscale('log')
plt.yscale('log')
plt.title('Both Axes in Log Scale (y = x^2)')
plt.grid(True, which="both", ls="--")
plt.show()
````

| Scale Type | Function                                              |
| ---------- | ----------------------------------------------------- |
| Linear     | Default                                               |
| Log        | `plt.xscale('log')`                                   |
| SymLog     | `plt.xscale('symlog')` (log for large, linear near 0) |
| Logit      | For probabilities (0 to 1) in logistic regression     |


- log scale is not ideal with data with zero or negative value
- label axes clearly when using log scales


| Feature             | Use                      |
| ------------------- | ------------------------ |
| `plt.xscale('log')` | Logarithmic X-axis       |
| `plt.yscale('log')` | Logarithmic Y-axis       |
| `np.logspace()`     | Log-scaled arrays        |
| `symlog` / `logit`  | Other useful scale types |
