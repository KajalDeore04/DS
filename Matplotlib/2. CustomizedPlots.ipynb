# Customization

```python
import matplotlib.pyplot as plt

plt.title("Title Name", fontsize=14, color='purple', loc='left')
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15,20, 25, 30]
y2 = [10, 20, 30, 40, 50]



plt.plot(x, y1, 
         label="Product A", 
         linestyle=':', 
         marker="s", 
         linewidth=2, 
         markersize=6)
plt.plot(x, y2, 
         label="Product B", 
         linestyle='-.', 
         marker="^", 
         linewidth=10, 
         markersize=10)

plt.legend(loc='upper left') # upper right, lower center to plot multiple lines
```

- colors can be - named, short codes(r,g,b), hex codes
- lines styles = -, --, :, -.
- markers = o, s, ^, *

## GridLines

```python
x = [1, 2, 3, 4, 5]
y = [10, 15,20, 25, 30]

plt.plot(x,y)
plt.grid(True,
         color='pink',
         linestyle='--',
         linewidth=1.5)
```

## Ticks

```python
x = [1, 2, 3, 4, 5]
y = [10, 15,20, 25, 30]

plt.plot(x, y)
plt.grid(True)

plt.xticks([0, 1, 2, 3], rotation=45)
plt.yticks([10, 20, 30], ['Low', 'Med', 'High'])

```

## Subplots and Layouts
- to display multiple plots in one figure, customize their layouts.

```python
# plt.subplot(2,2,1)   # create grid of plots with nrows, ncols, index(which plot position usually 1)

plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot([1, 2, 3], [1, 2, 3])
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot([1, 2, 3], [1, 0, -1])
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot([1, 2, 3], [1, 2, 1])
plt.title('Plot 4')

plt.tight_layout()  # Avoid overlapping
plt.show()
```

```python
# cleaner and flexible way

fig, axs = plt.subplots(2, 2, figsize=(10, 4))  # 2x2 grid of subplots

axs[0, 0].plot([1, 2, 3], [1, 4, 9])

axs[0, 1].bar([1, 2, 3], [3, 2, 5])

axs[1, 0].scatter([1, 2, 3], [3, 2, 1])

axs[1, 1].hist([1, 1, 2, 3, 4, 4, 4])

plt.tight_layout()
plt.show()
```

## Examples

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [1000, 1200, 1300, 900, 1500]

plt.plot(months, sales, marker='o', color='green', linestyle='--', label="Sales")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

labels = ['Q1', 'Q2', 'Q3', 'Q4']
productA = [100, 150, 200, 250]
productB = [80, 120, 160, 220]

x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, productA, width, label='Product A')
plt.bar(x + width/2, productB, width, label='Product B')

plt.xticks(x, labels)
plt.xlabel("Quarter")
plt.ylabel("Revenue")
plt.title("Quarterly Revenue by Product")
plt.legend()
plt.show()
```

```python
groupA = np.random.normal(60, 8, 100)
groupB = np.random.normal(70, 10, 100)

plt.hist(groupA, bins=10, alpha=0.6, label='Group A', color='blue')
plt.hist(groupB, bins=10, alpha=0.6, label='Group B', color='green')
plt.legend()
plt.title("Group Comparison")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```

```python
groupA = np.random.normal(70, 10, 100)
groupB = np.random.normal(80, 15, 100)
groupC = np.random.normal(60, 5, 100)

data = [groupA, groupB, groupC]

plt.boxplot(data, labels=['Group A', 'Group B', 'Group C'], patch_artist=True)
plt.title("Group Comparison")
plt.ylabel("Scores")
plt.grid(axis='y')
plt.show()
```

```python
x1 = np.random.rand(50)*50
y1 = x1 * 1.5 + np.random.normal(0, 5, 50)

x2 = np.random.rand(50)*50
y2 = x2 * 0.7 + np.random.normal(0, 5, 50)

plt.scatter(x1, y1, color='blue', label='Group A')
plt.scatter(x2, y2, color='orange', label='Group B')

plt.title("Comparison Between Two Groups")
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]

explode = (0.1, 0, 0, 0)  # "explode" the first slice

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode,
    shadow=True,
    colors=['#ff9999','#66b3ff','#99ff99','#ffcc99']
)
plt.title("Fruit Sales Breakdown")
plt.axis('equal')  # Equal aspect ratio for a perfect circle
plt.show()

```
