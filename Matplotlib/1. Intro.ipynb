# Matplotlib
- python lib to create visualization from data
- makes it easy to understand numbers at a glance
- | Task                            | Visual Aid   | Example               |
  | ------------------------------- | ------------ | --------------------- |
  | Understand patterns             | Line Plot    | Stock price over time |
  | Compare groups                  | Bar Chart    | Sales by region       |
  | Show distributions              | Histogram    | Age distribution      |
  | Identify relationships          | Scatter Plot | Height vs weight      |
  | Spot outliers or unusual points | Box Plot     | Salary outliers       |

## Steps involved:
1. Prepare figure - ```plt.figure() / plt.subploats()```
2. Draw the data (bars, lines, dots) - ```plt.plot(), plt.bar() etc```
3. Decorate - ```title, labels, legend```

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]     # x-axis values
y = [2, 4, 6, 8, 10]    # y-axis values

plt.plot(x,y)  # draw line plot
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

- matplot is foundation of other libraries like:

# Graphs

## 1. Line Plot
- ```plt.plot```
- to track changes over time
- to see trends 
- works best for continuous num data
- e.g. temp throughout the day

"Draw a line through the dots to see the journey"

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]         # Time (days)
y = [100, 120, 115, 130, 140]  # Stock Price

plt.plot(x, y)
plt.title("Stock Price Over Days")
plt.xlabel("Day")
plt.ylabel("Price")
plt.show()
````

## 2. Barchart
- ```plt.bar```
- to compare values of different categories
- shows who is bigger/smaller in comparision
- e.g. Sales of different products, marks in subjects

"How tall is each categories bar"

````python
x = ['Math', 'Science', 'History']
y = [80, 95, 70]

plt.bar(x, y)
plt.title("Marks in Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
````

## 3. Histogram
- ```plt.hist```
- to understand the distribution of num col
- to see frequence 
- e.g. how many people scored between 0-10, 10-20

"Buckets of values - to see how data is spread out"

````python
marks = [55, 67, 45, 78, 89, 45, 67, 56, 89, 90, 77, 80]

plt.hist(marks, bins=5) # bins = number of buckets
plt.title("Marks Distribution")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.show()
````

## 4. Scatter Plot
- ```plt.scatter```
- to see relation between two vars
- to detect pattern, trend, cluster
- ideal for correlation
- e.g. height vs weight

"Plotting dots on a map - each point = one person/data row"

````python
study_hours = [1, 2, 3, 4, 5, 6]
scores = [35, 50, 60, 65, 80, 85]

plt.scatter(study_hours, scores)
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
````

## 5. Pie Chart
- ```plt.pie```
- to show percentage / proportions
- works with parts of a whole
- e.g. market share of companies

"A pizza divided into slices - to show composition"

````python
labels = ['Rent', 'Food', 'Transport', 'Savings']
sizes = [40, 25, 15, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Monthly Expense Distribution")
plt.show()
````

## 6. Box Plot
- ```plt.boxplot```
- to spot outliers
- to understand range, median, quartiles
- to summarize a distribution in 5 numbers
- e.g. Salary distribution

"A box showing middle 50% and lines showing extremes - for data health check"

````python
data = [55, 45, 67, 78, 90, 120, 30, 110, 85]
plt.boxplot(data)
plt.title("Boxplot of Salaries")
plt.show()
````

| Data Type                    | You Want To...           | Use This Chart     |
| ---------------------------- | ------------------------ | ------------------ |
| Numerical (1 column)         | See distribution         | Histogram, Boxplot |
| Numerical (2 columns)        | Compare/change over time | Line Plot          |
| Categorical vs Numerical     | Compare categories       | Bar Chart          |
| Numerical vs Numerical       | See relationship         | Scatter Plot       |
| Categorical (parts of total) | See % share              | Pie Chart          |

## anatomy

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

fig, ax = plt.subplots()

# ax.plot(x,y)

ax.set_title("Simple line plot")

ax.set_xlabel("Days")
ax.set_ylabel("Scores")

ax.grid(True)

ax.set_xticks([1,2,3,4,5])
ax.set_yticks([10,15,20,25,30])

ax.plot(x,y, color='green', linestyle='--', marker='o', label="Score")

plt.savefig("trial.png")  # saves it as an image, pdf, .svg, .jpg
plt.show()
````

| Part           | Code                        | Purpose                                           |
| -------------- | --------------------------- | ------------------------------------------------- |
| Title          | `ax.set_title("...")`       | Describes what the graph is about                 |
| X Label        | `ax.set_xlabel("...")`      | Name for x-axis                                   |
| Y Label        | `ax.set_ylabel("...")`      | Name for y-axis                                   |
| Grid           | `ax.grid(True)`             | Adds dotted lines behind the plot for readability |
| Ticks          | `ax.set_xticks([...])`      | Customize where the numbers appear on axis        |
| Plot Line      | `ax.plot(x, y)`             | Draws the graph                                   |
| Legend         | `ax.legend()`               | Shows label for the line                          |
| Multiple Lines | Use multiple `ax.plot(...)` | You can draw many lines on same plot              |
